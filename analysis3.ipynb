```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import pi


def plot_radar_for_column(column_name):
    monthly_averages = df.groupby(['Year', 'Month'])[column_name].mean().unstack(level=0)
    
    normalized_data = (monthly_averages - monthly_averages.min()) / (monthly_averages.max() - monthly_averages.min())
    
    num_vars = len(normalized_data.index)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]  # Complete the loop
    
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    
    for year in normalized_data.columns:
        values = normalized_data[year].values.tolist()
        values += values[:1]  # Complete the loop
        ax.plot(angles, values, label=year)
        ax.fill(angles, values, alpha=0.1)
    
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    
    plt.xticks(angles[:-1], [str(month) for month in range(1, 13)])
    plt.yticks([0.2, 0.4, 0.6, 0.8], ["20%", "40%", "60%", "80%"], color="grey", size=12)
    plt.title(column_name, size=20, color="blue", y=1.1)
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
    
    plt.show()
```
